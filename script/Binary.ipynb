```python
# Imports & set path
import numpy as np
import pandas as pd
from pathlib import Path
import time, os

IN_URL = "https://data.austintexas.gov/resource/wter-evkm.csv"
OUT_URL = "https://data.austintexas.gov/resource/9t4d-g238.csv"

out_dir = Path("../raw_data")
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
# Read in 1000 rows of the data
#df_in = pd.read_csv(IN_URL, nrows=1000)
#df_out = pd.read_csv(OUT_URL, nrows=1000)

# Read in full data
# It doesn't work well with setting URL="...?$limit=...", so I used ChatGPT to read in the full data page by page
def fetch_socrata_to_csv(base_url, out_csv, limit=50000, max_rows=None):
    """
    pagination from Socrata to out_csv
    - Resumable download: If out_csv already exists, resume writing from the current row count
    """
    start_offset = 0
    if Path(out_csv).exists():
        try:
            start_offset = sum(1 for _ in open(out_csv, "r", encoding="urf-8")) -1
            if start_offset < 0:
                start_offset =0
            print(f"[Resume] Found existing {out_csv}, rows={start_offset}. Will continue from offset={start_offset}.")
        except Exception as e:
            print("Resume check failed:", e)
            start_offset = 0

    offset = start_offset
    total_written = start_offset
    t0 = time.time()
    header_written = Path(out_csv).exists() and start_offset > 0

    while True:
        url = f"{base_url}?$limit={limit}&$offset={offset}"
        try:
            chunk = pd.read_csv(url)
        except Exception as e:
            print(f"\nError fetching offset={offset}: {e}")
            print("WIll retry in 5s ...")
            time.sleep(5)
            continue

        n = len(chunk)
        if n == 0:
            break

        chunk.to_csv(out_csv, mode="a" if header_written else "w", index=False, header=not header_written)
        header_written = True

        offset += n
        total_written += n

        dt = time.time() - t0
        speed = total_written / max(dt,1)
        print(f"\rFetched & saved: {total_written} rows (last chunk={n}, ~{speed:.1f} rows/s)", end="")

        if max_rows is not None and total_written >= max_rows:
            print("\nReached max_rows limit; stopping early.")
            break

        if n < limit:
            break

    print("\nDone ->", out_csv)
    return total_written

in_csv = out_dir / "intakes.csv"
out_csv = out_dir / "outcomes.csv"

rows_in = fetch_socrata_to_csv(IN_URL, in_csv, limit=50000)
rows_out = fetch_socrata_to_csv(OUT_URL, out_csv, limit=50000)
print("Rows (intakes):", rows_in)
print("Rows (outcomes):", rows_out)

df_in = pd.read_csv(in_csv)
df_out = pd.read_csv(out_csv)
```

```python
# data cleaning
# datetime
# 1) convert time to UTC - then naive
def to_utc_naive(series):
    ts = pd.to_datetime(series, errors='coerce', utc=True)
    return ts.dt.tz_convert('UTC').dt.tz_localize(None)
    
# for Intakes: mainly use datetime (floating Timestamp) then datetime2
if "datetime" not in df_in.columns:
    raise ValueError("Intakes: missing 'datetime' column.")
df_in["intake_ts"] = to_utc_naive(df_in["datetime"])

# if there is still NaT, try using datetime2 (floating Timestamp), assume the first day of each month
if "datetime2" in df_in.columns:
    missing = df_in["intake_ts"].isna() & df_in["datetime2"].notna()
    if missing.any():
        alt = to_utc_naive(df_in.loc[missing, "datetime2"])
        still = alt.isna()
        if still.any():
            alt.loc[still] = to_utc_naive(
                df_in.loc[missing, "datetime2"].astype(str).str.slic(0,7) + "-01"
            ).loc[still]
        df_in.loc[missing, "intake_ts"] = alt

# for Outcomes: mainly use datetime: (Text) then monthyear
if "datetime" not in df_out.columns:
    raise ValueError("Outcomes: missing 'datetime' column.")
df_out["outcome_ts"] = to_utc_naive(df_out["datetime"])

# if there is still NaT, use monthyear (Text), assume the first day of each month
if "monthyear" in df_out.columns:
    missing = df_out["outcome_ts"].isna() & df_out["monthyear"].notna()
    if missing.any():
        m = pd.to_datetime(df_out.loc[missing, "monthyear"], format='%m-%Y', errors='coerce', utc=True)
        df_out.loc[missing, "outcome_ts"] = m.dt.tz_convert("UTC").dt.tz_localize(None)

# 2) convert age to days
def age_to_days(s):
    if pd.isna(s):
        return np.nan
    s = str(s).strip().lower()
    parts = s.split()
    if len(parts) < 2:
        return np.nan

    try:
        n = float(parts[0])
    except:
        return np.nan
    unit = parts[1]
    if "day" in unit: return n
    if "week" in unit: return n * 7
    if "month" in unit: return n * 30
    if "year" in unit: return n * 365
    return np.nan

if "age_upon_intake" in df_in.columns:
    df_in["age_upon_intake_days"] = df_in["age_upon_intake"].map(age_to_days)
if "age_upon_outcome" in df_out.columns:
    df_out["age_upon_outcome_days"] = df_out["age_upon_outcome"].map(age_to_days)

# 3) basic checks
print("Intakes: intake_ts non-null =", 1 - df_in["intake_ts"].isna().mean())
print("Outcomes: outcome_ts non-null =", 1 - df_out["outcome_ts"].isna().mean())

# 4)
aid = "animal_id" if "animal_id" in df_in.columns else ("animalid" if "animalid" in df_in.columns else None)
if aid is None:
    raise ValueError("Can't find column animal_id / animalid.")
time_in, time_out = "intake_ts", "outcome_ts"

# 5) fast check age and time
display(df_in[[aid, "intake_ts", "age_upon_intake", "age_upon_intake_days"]].head())
display(df_out[[aid, "outcome_ts", "age_upon_outcome", "age_upon_outcome_days"]].head())
```

```python
# === C) 合并时间线 + 组内 bfill （修正版，避免 KeyError）===
import pandas as pd
import numpy as np
from pathlib import Path

aid      = "animal_id"
time_in  = "intake_ts"   # 已在 B 中生成
time_out = "outcome_ts"  # 已在 B 中生成

# 1) 只保留需要的列（可按需增减）
in_keep = [c for c in [
    aid, time_in,
    "name","found_location","intake_type","intake_condition","animal_type",
    "sex_upon_intake","age_upon_intake","age_upon_intake_days","breed","color"
] if c in df_in.columns]

out_keep = [c for c in [
    aid, time_out,
    "outcome_type","outcome_subtype","animal_type","sex_upon_outcome",
    "age_upon_outcome","age_upon_outcome_days","breed","color"
] if c in df_out.columns]

df_in_  = df_in[in_keep].copy()
df_out_ = df_out[out_keep].copy()

# 2) 关键键统一 dtype
df_in_[aid]   = df_in_[aid].astype(str)
df_out_[aid]  = df_out_[aid].astype(str)

df_in_[time_in]   = pd.to_datetime(df_in_[time_in],  errors="coerce", utc=True).dt.tz_convert("UTC").dt.tz_localize(None)
df_out_[time_out] = pd.to_datetime(df_out_[time_out], errors="coerce", utc=True).dt.tz_convert("UTC").dt.tz_localize(None)

# 丢掉关键键缺失
df_in_  = df_in_.dropna(subset=[aid, time_in]).copy()
df_out_ = df_out_.dropna(subset=[aid, time_out]).copy()

# 3) 统一“事件时间列”为 event_ts；outcome 侧加前缀 out_
df_in_  = df_in_.rename(columns={time_in: "event_ts"})
df_out_ = df_out_.rename(columns={time_out: "event_ts"})
df_out_ = df_out_.rename(columns={c: f"out_{c}" for c in df_out_.columns if c not in [aid, "event_ts"]})

# 在 intake 表里为所有 out_* 列占位（否则 concat 后会出现缺列）
for c in [c for c in df_out_.columns if c.startswith("out_")]:
    if c not in df_in_.columns:
        df_in_[c] = np.nan

# 4) 合成“时间线”并标记来源（用列对齐避免 KeyError）
df_in_["__kind__"]  = "intake"
df_out_["__kind__"] = "outcome"

# 先取两边列的并集（包含 event_ts / __kind__ / aid）
common_cols = sorted(set(df_in_.columns) | set(df_out_.columns))
union_cols = [aid, "event_ts", "__kind__"] + [c for c in common_cols if c not in [aid, "event_ts", "__kind__"]]

# 关键：对齐两边的列，缺的列自动补 NaN，避免 df_out_[cols] 的 KeyError
df_in_aligned  = df_in_.reindex(columns=union_cols)
df_out_aligned = df_out_.reindex(columns=union_cols)

timeline = pd.concat([df_in_aligned, df_out_aligned], ignore_index=True)
timeline.sort_values([aid, "event_ts"], inplace=True, kind="mergesort")
timeline.reset_index(drop=True, inplace=True)

# 5) 先创建 outcome 时间的“临时列”，再确定要 bfill 的列集合
timeline["outcome_ts_tmp"] = np.where(timeline["__kind__"]=="outcome", timeline["event_ts"], pd.NaT)
out_cols = [c for c in timeline.columns if c.startswith("out_")] + ["outcome_ts_tmp"]

# 6) 组内 bfill：让每条 intake 拿到“之后最近”的 outcome
# 对每个 out_* 列分别按 animal_id 组内 bfill（向下填充“之后最近”的 outcome）
for c in out_cols:
    timeline[c] = timeline.groupby(aid, group_keys=False)[c].bfill()

# 7) 取回 intake 行，重命名列
df_pair = timeline[timeline["__kind__"]=="intake"].copy()
df_pair.rename(columns={"event_ts":"intake_ts", "outcome_ts_tmp":"out_outcome_ts"}, inplace=True)

# 8) 收容时长 & 负值剔除
# 先确保两列都是 pandas 的 datetime64[ns]（naive）
df_pair["intake_ts"] = pd.to_datetime(df_pair["intake_ts"], errors="coerce")
df_pair["out_outcome_ts"] = pd.to_datetime(df_pair["out_outcome_ts"], errors="coerce")

# 只对两边都有时间戳的行计算（避免 NaT 运算）
mask = df_pair["intake_ts"].notna() & df_pair["out_outcome_ts"].notna()

# 用 numpy 的 datetime64 做减法，再转成“天”
# （.values → 同为 ndarray；/ np.timedelta64(1,'D') → 单位天的浮点数）
df_pair.loc[mask, "shelter_days"] = (
    (df_pair.loc[mask, "out_outcome_ts"].values - df_pair.loc[mask, "intake_ts"].values)
    / np.timedelta64(1, "D")
)

# 对没匹配 outcome 的行留空
df_pair.loc[~mask, "shelter_days"] = np.nan

# 极少数脏数据可能导致负值，直接剔除
neg = (df_pair["shelter_days"] < 0).sum(skipna=True)
if neg:
    print(f"Found {neg} rows with shelter_days < 0; excluded.")
    df_pair = df_pair[(df_pair["shelter_days"].isna()) | (df_pair["shelter_days"] >= 0)].copy()

# 9) 未匹配掩码 & 建模集
unmatched_mask = df_pair["out_outcome_ts"].isna()
print("Matched ratio:", f"{1 - unmatched_mask.mean():.2%}")
df_pair_model = df_pair[~unmatched_mask].copy()

print("df_pair shape:", df_pair.shape, "| df_pair_model shape:", df_pair_model.shape)
display(df_pair[[aid, "intake_ts", "out_outcome_ts", "shelter_days"]].head(10))
```

```python
# Labeling and Category Distribution Check
out_type_col = next((c for c in df_pair.columns if c.endswith("outcome_type")), None)
assert out_type_col is not None, "Can't find the outcome_type column in df_pair."

# Define mapping rules
LIVE_TYPES = {"Adoption", "Return to Owner", "Transfer", "Relocate"}
NON_LIVE_TYPES = {"Euthanasia", "Died", "Disposal", "Missing", "Rto-Euthanasia"}

def map_live(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    if x in LIVE_TYPES:
        return 1
    if x in NON_LIVE_TYPES:
        return 0
    return np.nan

# labeling on df_pair and df_pair_model
df_pair["live_outcome"] = df_pair[out_type_col].map(map_live)
df_pair_model["live_outcome"] = df_pair_model[out_type_col].map(map_live)

# Population distribution & Missing proportion
def label_summary(frame, name):
    total = len(frame)
    n_pos = int((frame["live_outcome"] == 1).sum())
    n_neg = int((frame["live_outcome"] == 0).sum())
    n_nan = int((frame["live_outcome"].isna().sum()))
    print(f"\n=== Label Summary ({name}) ===")
    print(f"Total rows: {total}")
    print(f"Positive (live=1): {n_pos} ({n_pos/total:.2%})")
    print(f"Negative (live=0): {n_neg} ({n_neg/total:.2%})")
    print(f"Missing labels: {n_nan} ({n_nan/total:.2%})")
    return pd.DataFrame([{
    "dataset": name,
    "total": total,
    "positive": n_pos,
    "negative": n_neg,
    "nan": n_nan,
    "pos_rate": n_pos/total if total else np.nan,
    "nan_rate": n_nan/total if total else np.nan
    }])

sum_all = label_summary(df_pair, "df_pair (matched + unmatched)")
sum_model = label_summary(df_pair_model, "df_pair_model (matched only)")
label_stats = pd.concat([sum_all, sum_model], ignore_index=True)

# for modeling data only, remove missing labels
df_pair_model = df_pair_model.dropna(subset=["live_outcome"]).copy()
print(f"\nRows kept for modeling after dropping missing labels: {len(df_pair_model)}")
```

```python
# live rate by intake_type / animal_type
def rate_table(frame, key):
    if key not in frame.columns:
        return None
    t = frame.groupby(key)["live_outcome"].agg(["mean","count"]).sort_values("count", ascending=False)
    t.rename(columns={"mean":"live_rate", "count":"n"}, inplace=True)
    return t

rt_intake = rate_table(df_pair_model, "intake_type")
rt_animal = rate_table(df_pair_model, "animal_type")

if rt_intake is not None:
    print("\nTop intake_type by count (with live_rate):")
    display(rt_intake.head(10))
if rt_animal is not None:
    print("\nanimal_type live_rate:")
    display(rt_animal)
```

```python
# EDA-visualization
import matplotlib.pyplot as plt

eda = df_pair_model.dropna(subset = ["live_outcome"]).copy()
eda["live_outcome"] = eda["live_outcome"].astype(int)

fig_dir = (Path("..") if (Path("..")/"raw_data").exists() else Path(".")) / "figures"
fig_dir.mkdir(parents=True, exist_ok=True)
print("EDA rows:", len(eda))

def save_show_plt(figpath):
    plt.tight_layout()
    plt.savefig(figpath, dpi=200)
    plt.show()
    print("Saved:", figpath)

def exists_nonempty(cols):
    return all([(c in eda.columns) and (eda[c].notna().any()) for c in cols])

# figure 1: age_in vs live_outcome
if exists_nonempty(["age_upon_intake_days", "live_outcome"]):
    g0 = eda.loc[eda["live_outcome"]==0, "age_upon_intake_days"].dropna()
    g1 = eda.loc[eda["live_outcome"]==1, "age_upon_intake_days"].dropna()
    if len(g0)>5 and len(g1)>5:
        plt.figure(figsize=(5,3))
        plt.boxplot([g0,g1], labels=["non-live", "live"], showfliers=False)
        plt.ylabel("Age upon intake (days)")
        plt.title("Age at intake vs outcome")
        save_show_plt(fig_dir / "box_age_vs_outcome.png")
    else:
        print("Skip box_age_vs_outcome: not enough data in one of the groups.")
else:
    print("Skip box_age_vs_outcome: required columns missing -> age_upon_intake_days or live_outcome.")

# figure 2: intake_type vs live_outcome rate (top-8 by count)
if exists_nonempty(["intake_types", "live_outcome"]):
    tab = eda.groupby("intaake_type")["live_outcome"].agg(["mean","count"]).sort_values("count", ascending=false).head(8)
    if len(tab) >= 2:
        x = np.arange(len(tab))
        plt.figure(figsize=(7,4))
        plt.bar(x, tab["mean"].values)
        plt.xticks(x, tab.index, rotation=30, ha="right")
        plt.ylim(0,1)
        plt.ylabel("Live outcome rate")
        plt.title("Intake type vs live rate (Top-8 by count)")
        save_show_plt(fig_dir / "bar_intake_type_live_rate.png")
    else:
        print("Skip bar_intake_type_live_rate: not enough intake_type categories.")
else:
    print("Skip bar_intake_type_live_rate: 'intake_type' not found or empty.")

# figure 3: Top-10 breeds' live rates
if exists_nonempty(["breed", "live_outcome"]):
    topk = eda["breed"].value_counts().head(10).index.tolist()
    sub = eda[eda["breed"].isin(topk)].copy()
    if len(sub) > 0:
        tab = sub.groupby("breed")["live_outcome"].mean().sort_values(ascending=False)
        x = np.arange(len(tab))
        plt.figure(figsize=(8,4))
        plt.bar(x, tab.values)
        plt.xticks(x, tab.index, rotation=30, ha="right")
        plt.ylim(0,1)
        plt.ylabel("Live outcome rate")
        plt.title("Top-10 breeds by frequency: live rate")
        save_show_plt(fig_dir / "bar_top_breeds_live_rate.png")
    else:
        print("Skip bar_top_breeds_live_rate: no breeds after filtering.")
else:
    print("Skip bar_top_breeds_live_rate: 'breed' not found or empty.")

# figure 4: shelter_days distribution
if "shelter_days" in eda.columns and eda["shelter_days"].notna().any():
    sd = eda["shelter_days"].dropna()
    if len(sd) >= 10:
        plt.figure(figsize=(7,4))
        plt.hist(sd, bins="auto")
        plt.xlabel("Shelter days")
        plt.ylabel("Count")
        plt.title("Distribution of shelter stay length")
        save_show_plt(fig_dir / "hist_shelter_days.png")
    else:
        print("Skip hist_shelter_days: not enough non-null values.")
else:
    print("Skip hist_shelter_days: 'shelter_days' missing or empty (only exists for matched outcomes).")

# figure 5: Outcome age distribution
out_age_col = next((c for c in eda.columns if c.endswith("age_upon_outcome_days")), None)
if out_age_col is not None and eda[out_age_col].notna().any():
    x0 = eda.loc[eda["live_outcome"]==0, out_age_col].dropna()
    x1 = eda.loc[eda["live_outcome"]==1, out_age_col].dropna()
    if len(x0)>5 and len(x1)>5:
        hi = np.nanpercentile(eda[out_age_col], 99)
        hi = max(hi, 3650) # at least show 10 years
        plt.figure(figsize=(7,4))
        plt.hist(x0, bins=40, range=(0, hi), alpha=0.6, label="non-live")
        plt.hist(x1, bins=40, range=(0, hi), alpha=0.6, label="live")
        plt.xlabel("Age upon outcome (days)")
        plt.ylabel("Count")
        plt.title("Outcome age distribution")
        plt.legend()
        save_show_plt(fig_dir / "hist_age_upon_outcome_by_live.png")
    else:
        print("Skip hist_age_upon_outcome_by_live: not enough data in one of the groups.")
else:
    print("Skip hist_age_upon_outcome_by_live: 'out_age_upon_outcome_days' not available.")

# figure 6: intake_age vs shelter_days
if exists_nonempty(["age_upon_intake_days"]) and ("shelter_days" in eda.columns) and eda["shelter_days"].notna().any():
    sub = eda[["age_upon_intake_days", "shelter_days"]].dropna().copy()
    if len(sub) > 20:
        if len(sub) > 5000:
            sub = sub.sample(5000, random_state=0)
        plt.figure(figsize=(6,4))
        plt.scatter(sub["age_upon_intake_days"], sub["shelter_days"], s=8, alpha=0.4)
        plt.xlabel("Age upon intake (days)")
        plt.ylabel("Shelter days")
        plt.title("Intake age vs shelter stay")
        save_show_plt(fig_dir / "scatter_intake_age_vs_shelter_days.png")
    else:
        print("Skip scatter_intake_age_vs_shelter_days: not enough paired points.")
else:
    print("Skip scatter_intake_age_vs_shelter_days: required columns missing or empty.")
```

```python
# feature selection
target_col = "live_outcome"
group_col = "animal_id"

# categorical variables
cat_cols = [c for c in [
    "animal_type",
    "breed",
    "color",
    "intake_type",
    "intake_condition",
    "sex_upon_intake",
    "intake_subtype"
] if c in df_pair_model.columns]

# numerical variables
num_cols = [c for c in [
    "age_upon_intake_days",
    # "shelter_days" don't use it since it will cause data leakage (it only exists after the outcome is known) (same as out_age_upon_outcome_days)
] if c in df_pair_model.columns]

use_cols = cat_cols + num_cols + [target_col, group_col]
data = df_pair_model[use_cols].dropna(subset=[target_col, group_col]).copy()

X = data[cat_cols + num_cols]
y = data[target_col].astype(int)
groups = data[group_col]

print("X shape: ", X.shape)
print("y shape: ", y.shape)
print("Categorical: ", cat_cols)
print("numerical: ", num_cols)
y.value_counts(normalize=True).rename("proportion")
```

```python
# Group-based split: train/val/test = 70/15/15
from sklearn.model_selection import GroupShuffleSplit

gss1 = GroupShuffleSplit(n_splits=1, train_size=0.85, random_state=42)
trainval_idx, test_idx = next(gss1.split(X, y, groups=groups))

X_trainval, X_test = X.iloc[trainval_idx], X.iloc[test_idx]
y_trainval, y_test = y.iloc[trainval_idx], y.iloc[test_idx]
groups_trainval = groups.iloc[trainval_idx]

gss2 = GroupShuffleSplit(n_splits=1, train_size=0.8235, random_state=42)
train_idx, val_idx = next(gss2.split(X_trainval, y_trainval, groups=groups_trainval))

X_train, X_val = X_trainval.iloc[train_idx], X_trainval.iloc[val_idx]
y_train, y_val = y_trainval.iloc[train_idx], y_trainval.iloc[val_idx]

print("n_train:", len(X_train), "n_val:", len(X_val), "n_test:", len(X_test))
```

```python
# Preprocessing + Baseline Logistic Regression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

preprocessor = ColumnTransformer([
    ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False), cat_cols),
    ("scale", StandardScaler(), num_cols)
])

clf = Pipeline(steps = [
    ("prep", preprocessor),
    ("logreg", LogisticRegression(penalty='l2', max_iter=1000, random_state=0))
])

# only fit on train set
clf.fit(X_train, y_train)

# validation and test
val_pred = clf.predict(X_val)
test_pred = clf.predict(X_test)

val_score = accuracy_score(y_val, val_pred)
test_score = accuracy_score(y_test, test_pred)

print("Validation accuracy:", val_score)
print("Test accuracy:", test_score)
```

```python
# Confusion Matrix
print("\nValidation confusion matrix [[TN, FP], [FN, TP]]:")
print(confusion_matrix(y_val, val_pred))
print("\nTest confusion matrix [[TN, FP], [FN, TP]]:")
print(confusion_matrix(y_test, test_pred))
```

```python
# Feature numbers before and after preprocessing
ohe = clf.named_steps["prep"].named_transformers_["onehot"]
ohe_feature_names = ohe.get_feature_names_out(cat_cols) if len(cat_cols)>0 else np.array([])

n_features_before = X_train.shape[1]
n_features_after = len(ohe_feature_names) + len(num_cols)

print("Features before preprocessing:", n_features_before)
print("Features after preprocessing:", n_features_after)
pd.Series(ohe_feature_names).to_csv(inter_dir / "ohe_feature_names.csv", index=False)
```
